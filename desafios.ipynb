Desafio 3: Transcrição de uma molécula de DNA

```java
public class dna {
    public static void converteDNA(String DNA) {
        char RNA[];
        RNA = new char[15];
        for(int i =0; i<DNA.length();i++) {
            char c = DNA.charAt(i);
            if(c=='A') {       RNA[i]='U';
            } else if(c=='C') {RNA[i]='G';
            } else if(c=='G') {RNA[i]='C';
            } else if(c=='T') {RNA[i]='A';
            } else RNA[i]='X'; //indicativo de erro na decodificação
        }
        
        for(int j = 0; j<DNA.length();j++) {
            System.out.print(RNA[j]);
        }
    }
}    
```

```java
converteDNA("ACGTACGTACGTACG");
```

Desafio 4: Média salarial

```java
import java.util.Random;

public class salario {
    public static void main(String args[]) {
    Random random = new Random();
    
        int acum=0;
        double Salario[];
        Salario = new double[50];
        int cont=0;


        for (int i=0; i<50; i++){
    
            int N = random.nextInt((15000 - 2500) + 1) + 2500;
            acum=acum+N;
            //cria um vetor com os 50 salários
            Salario[i]=N;
        }
            for(int j=0; j<50;j++) {
            if(Salario[j]>=acum/50) {
                cont++;
            }
        }
        System.out.println("media:"+acum/50);
        System.out.println("recebem acima da media:"+cont);
}
}
```
