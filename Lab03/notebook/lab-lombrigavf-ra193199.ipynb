```java
public class AquarioLombriga {
int tamAquario;
int tamLombrig;
int posLombrig;
char lombriga[]; 

public AquarioLombriga(int A, int B, int C) {
        tamAquario = A;
        tamLombrig = B;
        posLombrig = C;
lombriga=new char[A];
}

public void imprimeLombriga(AquarioLombriga g) {
    
        for(int i=1;i<=g.tamLombrig;i++) 
            System.out.print(g.lombriga[i]);
}


public void imprimeAquario(int N) {

        for(int i=1;i<=N;i++) 
            System.out.print("#");

}

public static void Apresenta(AquarioLombriga g) {
        int atual=1;
        for(int j=1;j<=g.tamAquario;j++){
            if(atual != g.posLombrig) {g.imprimeAquario(1);
            atual++;
            } else {
                g.imprimeLombriga(g);
                atual=atual+g.tamLombrig;
                j=j+g.tamLombrig-1;
            }
        }System.out.println();
}
        

    public static void Crescer(AquarioLombriga g) {
        if(g.posLombrig+g.tamLombrig<=g.tamAquario) {
          g.tamLombrig=g.tamLombrig+1;
             int N=g.tamLombrig;
            if(g.lombriga[1] =='O') {
             g.lombriga[N]='@';	
              g.posLombrig=g.posLombrig+1;
             } else {
                g.lombriga[N-1]='@';
                g.lombriga[N]='O';
                g.posLombrig=g.posLombrig-1;
    }   
  }
}
     public static void Mover(AquarioLombriga g) {

        if(g.lombriga[1]=='O')g.posLombrig=g.posLombrig-1;
        else g.posLombrig=g.posLombrig+1;

    }

    public static void Virar(AquarioLombriga g) {
        for(int i =1;i<=g.tamLombrig;i++) {
            if(i==1)g.lombriga[i]='O';
            else g.lombriga[i]='@';
        }
    }

    public static void criaLombriga(AquarioLombriga g) {
        for(int i =1;i<=g.tamLombrig;i++) {
            if(i==g.tamLombrig)g.lombriga[i]='O';
            else g.lombriga[i]='@';
        }
    }
}

```
