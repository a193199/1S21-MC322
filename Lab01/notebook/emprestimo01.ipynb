```c
//continuação laboratório de empréstimo -> backup

#include <stdio.h>
void imprime_parcela (int N, int S, float j){
    while(N < 0){
        
        float Sf=S*(1+j);
        printf("Parcela %d: %.2f\n",N ,Sf);
        S=Sf;
        N--;    
    }
}

int main () {
    imprime_parcela(5,200,0.01);
    //int N = 5;
    //int S = 200;
    //float Sf = 0;       
    }
```

```c
//criar uma nova função em que exista uma memória dos parâmetros fixos do empréstimo
//método: criação de uma structure

//P representa a parcela desejada
//j representa a taxa de juros
//S representa o valor da parcela inicial
//N representa o número total de parcelas
#include <stdio.h>
typedef struct {
    int S; 
    float j;
} data;

void proximaParcela (int P, data A){

    if (P == 1){printf("%d\n",A.S);}
    else {
       
    int i = 0; //contador
    float Sf; //acumulador
    float Sf2 = A.S; //acumulador
        while(i < P-1){
            
            Sf=Sf2*(1+A.j);
            Sf2=Sf;
            i++;    
        } printf("%.2f\n",Sf2);
    }   
}

//o programa deve ser capaz de calcular as parcelas individualmente 
//a partir dos parâmetros atribuídos

int main () {
    data parametro;
    parametro.S=200;
    parametro.j=0.01;
    
    proximaParcela(5, parametro);       
    }

```

```c
//função que recebe apenas um parâmetro

#include <stdio.h>
typedef struct {
    int S; 
    float j;
    int pos;
    int N;
} data;

void proximaParcela (data A){

    if (A.pos == 1){printf("%d\n",A.S);}
    else {
       
    int i = 0; //contador
    float Sf; //acumulador
    float Sf2 = A.S; //acumulador
        while(i < A.pos-1){
            
            Sf=Sf2*(1+A.j);
            Sf2=Sf;
            i++;    
        } printf("%.2f\n",Sf2);
    }   
}



int main () {
    data parametro;
    parametro.S=200;
    parametro.j=0.01;
    parametro.N=5;

    
    int i=1;
    while (i <= parametro.N){
       parametro.pos=i; 
       proximaParcela(parametro);
       i++; 
    }
    
}
```

```c
//Estágio 5 -> contrução de uma estrutura capaz de calcular vários empréstimos
//ao mesmo tempo

//criar um vetor de struct para guardar as informações de multiplos variáveis

#include <stdio.h>
#include <string.h>
#include <stdlib.h>

typedef struct {
    int S;        //parcela inicial
    float j;      //juros
    int pos;      //posição que deve ser impressa
    int N;        //máximo de parcelas
} data;
data parametro[2];

void proximaParcela (int num, data A[]){
    if(A[num].pos > A[num].N){
        printf("Erro: Limite excedido\n");
    }    
    
    if (A[num].pos == 1){
        printf("%d\n", A[num].S);
    } else {
       
    int i = 0; //contador
    float Sf; //acumulador
    float Sf2 = A[num].S; //acumulador
        if(A[num].pos < A[num].N){
            while(i < (A[num].pos )){

                Sf=Sf2*(1+A[num].j);
                Sf2=Sf;
                i++;   
            }
            printf("Emprestimo:%d parcela %d eh %.2f\n\n",num,A[num].pos,Sf2);
        } 
    }   
}


int main () {
      
    parametro[0].S=600;
    parametro[0].j=0.01;
    parametro[0].N=5;
    //parametro[0].pos=5;
    
    parametro[1].S=300;
    parametro[1].j=0.05;
    parametro[1].N=8;
    //parametro[1].pos=5;
    
    parametro[2].S=200;
    parametro[2].j=0.01;
    parametro[2].N=5;
    //parametro[2].pos=5;
    
    int num=2;//identificador do empréstimo
    parametro[num].pos=4; //parcela desejada
    
    proximaParcela(num, parametro);
   
}
```
